```python
filename = 'finance.json'
```

```python
# imports
import json
from collections import namedtuple
from itertools import islice
```

```python
json_row = namedtuple('JsonRow', 'email contact name room job')
def data():
    with open(filename) as f:
        j = json.load(f)
        entries = j['feed']['entry']
        for entry in entries:
            email = entry['gsx$email']['$t']
            contact = entry['gsx$contact']['$t']
            job = entry['gsx$duty']['$t']
            name = entry['gsx$name']['$t']
            room = entry['gsx$room']['$t']
            yield json_row(email, contact, name, room, job)
        
```

```python
def data_list():
    return list(data())
```

```python
list(islice(data_list(), 4))
```

```python
rowtuple = namedtuple('row', 'firstName lastName middleName email job mobile roomContact, office')
def gen_rows():
    for _tuple in data():
        names = _tuple.name.split(' ')
        firstName = names[0]
        lastName = names[-1]
        middleName = None
        if len(names) == 3:
            middleName = names[1]
        yield rowtuple(firstName, lastName, middleName, _tuple.email, _tuple.job, None, _tuple.contact, _tuple.room)
```

```python
from datetime import datetime, date
d = date.today()
# print(datetime.fromtimestamp(1574757654476))
1574757628
print(datetime.now()) // returns current date and time
now = datetime.now()
print(datetime.timestamp(datetime.now()) * 1000) # returns timestamp converted to milliseconds
```

```python
def to_json():
    json_list = []
    for row in gen_rows():
        json_obj = {
            'firstName': row.firstName,
            'lastName': row.lastName,
            'middleName': row.middleName,
            'email': row.email,
            'job': row.job,
            'roomContact': row.roomContact,
            'office': row.office,
            'lastModified': datetime.timestamp(datetime.now()) * 1000 
        }
        json_list.append(json_obj)
    return json_list
```

```python
to_json()
```

```python
processed_filename = 'processed_finances.json'
with open(processed_filename, 'w') as f:
    json.dump(to_json(), f)
```
